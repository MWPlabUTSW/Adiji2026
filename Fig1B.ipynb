```python
import pandas as pd
import csv
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import metapredict as meta
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import scipy as scipy
from scipy import stats
```

```python
#Create dictionary of proteome with keys,values as ">" first line and seq, respectively
f = open('UP000000803_7227.fasta', 'r')
PrSeq = {}
for line in f:
    if line.startswith('>'):
        key = line[0:line.find('_')]
        #key = key.strip()
        key = key.replace('\n', '')
        seq = ''
    else:
        seq = seq + line
    #seq = seq.strip()
    seq = seq.replace('\n', '')
    PrSeq[key] = seq
```

```python
#Predict disordered domains and make a dictionary of proteins (key) and their disordered domains (value)
IDRs = {}
for Pr,Seq in PrSeq.items():
    T = {}
    eI = []
    t = []
    if len(Seq) >= 50 and Seq.count('X') == 0 and Seq.count('U') == 0:
        O = meta.predict_disorder_domains(Seq)
        I = O.disordered_domain_boundaries
        S = O.disordered_domains
        for reg in I:
            t.append([reg[0]+1,reg[1]])
        for i in range(len(t)):
            T[str(t[i])] = S[i]
    IDRs[Pr]=T
```

```python
lengths = []
fullsites = []
halfsites = []
fulldensity = []
halfdensity = []
for key in IDRs:
    for key1,value1 in IDRs[key].items():
        dns = []
        #calculate full site number ([S/T]Px[R/K])
        if len(value1) > 50:
            x=0
            for i in range(len(value1)-3):
                if (value1[i]=='S' or value1[i]=='T') and value1[i+1]=='P' and (value1[i+3]=='R' or value1[i+3]=='K'):
                    x = x+1
            fullsites.append(x)
            fulldensity.append(x/(len(value1)/100))
            #calculate half site number ([S/T]P)
            y=0
            for i in range(len(value1)-1):
                if (value1[i]=='S' or value1[i]=='T') and value1[i+1]=='P':
                    y = y+1
            halfsites.append(y)
            halfdensity.append(y/(len(value1)/100))

            lengths.append(len(value1))
```
