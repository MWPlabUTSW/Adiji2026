```python
#This code is used to count the number of CDK half and full sites
#plus CKII sites in Orc1 orthologs
WTSiteDensity = {}

for Prt,Seq in Seqs.items():
    
    FullSites = []
    HalfSites = []
    CKSites = []

    x = 0 #for full site counting
    y = 0 #for half site counting
    z = 0 #for CKII site counting
        
    #Counts CDK full sites and stores in x
    for i in range(len(Seq)-3):
        if (Seq[i]=='S' or Seq[i]=='T') and Seq[i+1]=='P' and (Seq[i+3]=='R' or Seq[i+3]=='K'):
            x=x+1
    FullSites.append(x/(len(Seq)/100))
        
    #Counts CDK half sites and stores in y
    for i in range(len(Seq)-2):
        if (Seq[i]=='S' or Seq[i]=='T') and Seq[i+1]=='P':
            y=y+1
    HalfSites.append(y/(len(Seq)/100))
        
    #Counts CKII sites and stores in z
    for i in range(len(Seq)-3):
        if (Seq[i]=='S' or Seq[i]=='T') and (Seq[i+3]=='D' or Seq[i+3]=='E'):
            z=z+1
    CKSites.append(z/len(Seq)/100)
    
    WTSites = [HalfSites,FullSites,CKSites]
    
    WTSiteDensity[Prt]=WTSites  
```

```python
for key,items in WTSiteDensity.items():
    print(key)
    print("WT CDK Half Sites:", items[0])
    print("WT CDK Half Sites:", items[1])
    print("WT CKII Sites:", items[2])
```

```python
#This code is used to count the number of CDK half and full sites
#plus CKII sites in 10,000 random scrambles of Orc1 orthologs
SiteDensity = {}

for Prt,Seq in Seqs.items():
    
    FullSites = []
    HalfSites = []
    CKSites = []
    
    #Generate 10,000 scrambles
    for p in range(10000):
        
        scrambled = ''.join(random.sample(Seq, len(Seq)))
        
        x = 0 #for full site counting
        y = 0 #for half site counting
        z = 0 #for CKII site counting
        
        #Counts CDK full sites and stores in x
        for i in range(len(scrambled)-3):
            if (scrambled[i]=='S' or scrambled[i]=='T') and scrambled[i+1]=='P' and (scrambled[i+3]=='R' or scrambled[i+3]=='K'):
                       x=x+1
        FullSites.append(x/(len(scrambled)/100))
        
        #Counts CDK half sites and stores in y
        for i in range(len(scrambled)-2):
            if (scrambled[i]=='S' or scrambled[i]=='T') and scrambled[i+1]=='P':
                       y=y+1
        HalfSites.append(y/(len(scrambled)/100))
        
        #Counts CKII sites and stores in z
        for i in range(len(scrambled)-3):
            if (scrambled[i]=='S' or scrambled[i]=='T') and (scrambled[i+3]=='D' or scrambled[i+3]=='E'):
                        z=z+1
        CKSites.append(z/len(scrambled)/100)
    
    Sites = [HalfSites,FullSites,CKSites]
    
    SiteDensity[Prt]=Sites  
```

```python
for key,items in SiteDensity.items():
    print(key)
    print("Random CDK Half Sites:", sum(items[0])/len(items[0]))
    print("Random CDK Half Sites:", sum(items[1])/len(items[1]))
    print("Random CKII Sites:", sum(items[2])/len(items[2]))
```
